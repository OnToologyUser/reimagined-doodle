```python
import csv
fields=['Name','Description','Category']
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver=webdriver.Chrome("/Users/HP/Downloads/chromedriver_win32/chromedriver.exe")
url="https://en.wikipedia.org/wiki/List_of_Indian_dishes"
driver.get(url)
page=driver.page_source
soup=BeautifulSoup(page,'html.parser')
```

```python
tag=soup.find('div',class_="mw-parser-output")
data=tag.find_all('table',class_="wikitable")
head=tag.find_all('span',class_='mw-headline')
k=1
for zone in data:
     rows=[]
     filename=head[k].text+".csv"
     tbody=zone.find('tbody')
     tr=tbody.find_all('tr')
     for i in tr:
         td=i.find_all('td')
         l=0
         column=[]
         for j in td:
             if(l==1):
                 l+=1
                 continue
             column.append(j.text)
             l+=1
         rows.append(column)
        #  print(rows)
     with open(filename,'w',newline='') as csvfile:
         writer=csv.writer(csvfile)
         writer.writerow(fields)
         writer.writerows(rows)
     k+=1
```
